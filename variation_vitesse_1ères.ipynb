```python
import numpy as np
from scipy import *
import matplotlib.pyplot as plt
%matplotlib inline

#nombre de points
N = 15

Dt=0.800

a=-2.70190239524
b=2.24850030182
c=0.00127582949017
dt=0.400
ech_v=0.3
t=np.linspace(0,Dt,N)
g=9.86
x=0-g/2/a*t
y=c+b*t+a*t**2
plt.plot(x,y,"bo",markersize=4)

for i in range(0,N):
    plt.text(x[i],y[i]+0.02,'M'+str(i),fontsize=8)
    # plt.pause(dt)
    
vx = [(x[i+1]-x[i-1])/(2*dt) for i in range(1,N-1)]
vy = [(y[i+1]-y[i-1])/(2*dt) for i in range(1,N-1)]

for i in range(1,N-1):
    plt.arrow(x[i],y[i],ech_v*vx[i-1],ech_v*vy[i-1],head_width=0.01,head_length=0.01,length_includes_head=True,color="r")
    plt.text(0.4,0.1,"vecteur vitesse",color="red")
    # plt.pause(dt)

for i in range(2,N-2):
    plt.arrow(x[i],y[i],vx[i]-vx[i-2],vy[i]-vy[i-2],head_width=0.01,head_length=0.01,length_includes_head=True,color="g")
    plt.text(0.4,0.05,"vecteur variation",color="green")
    # plt.pause(dt)

plt.plot(x,y,"bo",markersize=4)
plt.xlabel("x(m)")
plt.ylabel("y(m)")
plt.title("Trajectoire de la balle")
plt.grid()
plt.show()
```
